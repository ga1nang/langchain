```python
import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
os.chdir('e:\data_scien\course\Complete_genai_langchain_huggingface\learn\project_langchain')

load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "My nick name is Ga1nang"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
query_result_1024 = embeddings_1024.embed_query(text)
print(query_result_1024)
```

```python
len(query_result_1024)
```

```python
loader = TextLoader('data/speech.txt')
docs = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
#vector embedding and vector storeDB
db = Chroma.from_documents(final_docs, embeddings_1024)
db
```

```python
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_res = db.similarity_search(query)
print(retrieved_res)
```
